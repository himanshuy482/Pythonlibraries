# Seaborn is a data visualisation library which could be used inplace of matplotlib or with matplotlib and has several inbuilt plottypes

```python
import pandas as pd
import numpy as np
```

```python
# download in conda using command prompt
# To import seaborn
import seaborn as sns
# suppose if you want to display image after the cell executes without creating an figure instance use next command
%matplotlib inline
# To import matplotlib
import matplotlib.pyplot as plt

```

# I will be visualising the pokemon dataset using Seaborn

```python
# importing dataset
train = pd.read_csv('Pokemon.csv')
```

```python
train.head()
```

```python
train['Type 1'].value_counts()
```

# Categorical plots


```python
# barplot 
plt.figure(figsize = (12,3))
sns.barplot(x = 'Type 1',y = 'Attack',data = train)
```

```python
# countplot
# To count the frequency of categorical variables
plt.figure(figsize = (12,6))
sns.countplot(x = 'Type 1',data = train,saturation = 10,hue = 'Legendary')
# Try without using hue and understand the difference
```

```python
# Boxplot
#In a boxplot box shows the quartiles of a dataset while whiskers extend to show complete distribution
#Except for points that are determined outliers using a function called inter quartile range
plt.figure(figsize = (15,8))
sns.boxplot(x ='Type 2',y = 'Defense',data = train)
```

```python
# violin plot
# Unlike a box plot, in which all of the plot components correspond to actual datapoints,
#the violin plot features a kernel density estimation of the underlying distribution.
plt.figure(figsize = (14,8))
sns.violinplot(x ='Type 2',y = 'Defense',data = train,palette='rainbow')
```

```python
# Stripplot
#The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own,
#but it is also a good complement to a box or violin plot in cases where you want to show all observations.
plt.figure(figsize = (14,6))
sns.stripplot(x = 'Type 2',y = 'Attack',data = train)
```

```python
# swarmplot
# Similar to stripplot but points donot overlap in a swarmplot
plt.figure(figsize = (14,4))
sns.swarmplot(x = 'Type 1',y = 'Speed',hue = 'Legendary',data = train,palette='rainbow')
```

```python
# factorplot
#factorplot is the most general form of a categorical plot. It can take in a kind parameter to adjust the plot type.
sns.factorplot(x = 'Type 1',y = 'Attack',data = train,kind ='bar',aspect = 4,ci = 0.80)
```

# Distribution Plots
# Allows us to visualise the distribution of the data set
# distplot,jointplot,kdeplot,rugplot,pairplot

```python
# distplot
sns.distplot(train['Sp. Atk'],kde = False)
```

```python
# jointplot
# jointplot allows us to basically matchup two distplots for bivariate data
sns.jointplot(x = 'Sp. Atk',y = 'Attack',data = train,kind = 'scatter')
```

```python
# pairplot
sns.pairplot(train)
```

```python
#rugplot
# rugplot just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:
sns.rugplot(train['Speed'])
```

# Matrix plots
# heatmap

```python
# heatmap
# plots the correlation matrix
plt.figure(figsize = (14,4))
sns.heatmap(train.corr())
```
