# to remove columns

```python
train.drop('age of pokemon',inplace = True,axis = 1)
# note here axis = 1 denotes that we want to drop along column.
# inplace = True means that i am permanently removing this column from my dataframe and cannot modify it lateron
# this ll be more clear to you if you assign a new variable to the data frame and use .drop( ) method for both inplace = True and inplace = False 
```

# to select rows using loc[ ] ans iloc[ ] method

```python
train.loc[2]
# note  loc method works on labels on the index whereas iloc works on the  positions in the index so takes only integer values . 
```

```python
# conditional selection in pandas 
train['Defense']>50
# use & for intersection of two conditionals and | for union of two conditionals .
```

# to set our indexes to a different value inplace of integers 

```python
# here we index by'Name'
train.set_index('Name')
```

# groupby() method allows us to group rows of a dataframe together for a column name and help us to use aggregate functions on that new object

```python
pokes = train.groupby('Type 1')
pokes.mean()
```

#  How to deal with Missing values in dataframes 

```python
# we can use .drop() method to drop rows with any missing values in the dataframe or a column of a dataframe
train.dropna(axis = 0)
```

```python
# fillna() method fills missing values with none
train.fillna(value = 'None')
```
